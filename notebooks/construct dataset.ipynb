```python
import pandas as pd

from src.config import DEFAULT_DATA_FOLDER
from src.data.utils import save_as_excel
```

```python
architectural_posts = pd.read_excel(f"{DEFAULT_DATA_FOLDER}/output/architectural_posts_results.xlsx")
architectural_posts =  architectural_posts.drop(columns=['Purpose'])
architectural_posts.head(10)
```

```python
programming_posts = pd.read_excel(f"{DEFAULT_DATA_FOLDER}/output/programming_posts_results.xlsx")
programming_posts.head(10)
```

```python
combined_posts = pd.concat([architectural_posts, programming_posts], ignore_index=True)
combined_posts.head(10)
```

```python
print("Total architectural posts:", len(architectural_posts))
print("Total programming posts:", len(programming_posts))
print("Total combined posts:", len(combined_posts))
```

```python
save_as_excel(combined_posts, f"{DEFAULT_DATA_FOLDER}/output/combined_posts_results.xlsx")
print("Completed")
```
